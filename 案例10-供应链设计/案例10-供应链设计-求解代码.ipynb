```python
# 可信度
# 供应商的可信度Rs[i,t] = exp{-lambda_it * 12}
Rs =[ [0.006229869,0.047226505,0.011852695,0.018169698],
    [0.003763522,0.148021166,0.012555456,0.206139944],
    [0.145206269,0.236927759,0.124531074,0.006504897],
    [0.027853407,0.267135302,0.014781188,0.23579323],
    [0.01113569,0.018082693,0.136422308,0.252183067],
    [0.011967029,0.014995577,0.006599245,0.077801077],
    [0.07817542,0.066723371,0.106288307,0.003017882]
]

# 仓库的可信度Rw[t] = exp{-lambda_t * 12}
Rw = [5.4834E-05,2.80703E-05,2.59952E-05,2.46582E-05]

print(Rs[1][1])
```

```python
# 定义线性规划问题
from pulp import *
prob = LpProblem("supplyChain", LpMinimize)

# 创建变量
# X[ijt] Amount of jth product ordered to ith supplier in tth period,
X = LpVariable.dicts('X',[(i,j,t) for i in range(1,M + 1) for j in range(1,PR+1) for t in range(1,PE+1)],lowBound=0,cat = 'Continuous')
# XK[kjt] Amount of jth product dispatched by the central warehouse to wholesaler k in tth period
XK = LpVariable.dicts('XK',[(k,j,t) for k in range(1,KR+1) for j in range(1,PR+1) for t in range(1,PE+1)],lowBound=0,cat = 'Continuous')
# I[j,t] Inventory amount of jth product in tth period,
I = LpVariable.dicts('I',[(j,t) for j in range(1,PR+1) for t in range(0,PE+1)],cat = 'Continuous',lowBound=0)
# D[jt] Demand of the central warehouse for jth product in tth period,
D = LpVariable.dicts('D',[(j,t) for j in range(1,PR+1) for t in range(1,PE+1)],cat = 'Continuous',lowBound=0)
# L[jt] Inventory amount of jth product in tth period
L = LpVariable.dicts('L',[(j,t) for j in range(1,PR+1) for t in range(1,PE+1)],cat = 'Continuous',lowBound=0)
# LO'[jt] Lost sales amount of jth product at the end of tth period for the central warehouse
LO1 = LpVariable.dicts('LO1',[(k,j,t)for k in range(1,KR+1) for j in range(1,PR+1) for t in range(1,PE+1)],cat = 'Continuous',lowBound=0)
# OX[t] Number of temporary warehouses established in tth period
OX = LpVariable.dicts('OX',[t for t in range(1,PE+1)],cat='Integer',lowBound=0)
# Y[ijt] (Binary) Binary variable taking the value of 1 when ith supplier is selected to supply jth product in tth period; otherwise, it is equalto 0,
Y = LpVariable.dicts('Y',[(i,j,t) for i in range(1,M + 1) for j in range(1,PR+1) for t in range(1,PE+1)],cat = 'Binary')
# B[kjt](Binary) Binary variable taking the value of 1 when the demand of  wholesaler k is satisfied for jth product in tth period; otherwise, it is equal to 0.
B = LpVariable.dicts('B',[(k,j,t) for k in range(1,KR+1) for j in range(1,PR+1) for t in range(1,PE+1)],cat = 'Binary')

TotalCost = LpVariable("TotalCost", lowBound=0, cat='Continuous') 
TotalWeight = LpVariable("TotalWeight", lowBound=0, cat='Continuous')
TotalReliability = LpVariable("TotalReliability", lowBound=0, cat='Continuous')

PurchaseCost = LpVariable("PurchaseCost", lowBound=0, cat='Continuous') 
OrderCost = LpVariable("OrderCost", lowBound=0, cat='Continuous') 
HoldingCost = LpVariable("HoldingCost", lowBound=0, cat='Continuous') 
SellLossCost = LpVariable("SellLossCost", lowBound=0, cat='Continuous') 
TransportCost = LpVariable("TransportCost", lowBound=0, cat='Continuous') 
TempWarehouseCost = LpVariable("TempWarehouseCost", lowBound=0, cat='Continuous')
TotalSupplierWeight = LpVariable("TotalSupplierWeight", lowBound=0, cat='Continuous')

# 目标函数
prob += 0.5*TotalCost -0.3*TotalWeight -0.2 * TotalReliability

# for i in range(1,M+1) 
# for j in range(1,PR+1) 
# for t in range(1,PE+1)
# for k in range(1,KR+1)

# 约束条件
# 总成本约束
prob += TotalCost == PurchaseCost + HoldingCost + OrderCost + SellLossCost + TransportCost + TempWarehouseCost
prob += PurchaseCost == lpSum(P[i-1][j-1] * X[i,j,t] for j in range(1,PR+1) for i in range(1,M+1) for t in range(1,PE+1))
prob += HoldingCost == lpSum(h[j-1] * I[j,t] for j in range(1,PR+1) for t in range(1,PE+1) )
prob += OrderCost == lpSum(O[i-1][j-1] * Y[i,j,t] for i in range(1,M+1) for j in range(1,PR+1) for t in range(1,PE+1))
prob += SellLossCost == lpSum(LW[j-1] * LO1[k,j,t] for k in range(1,KR+1) for j in range(1,PR+1) for t in range(1,PE+1))
prob += TransportCost == lpSum(DO[k-1] * XK[k,j,t] for k in range(1,KR+1) for j in range(1,PR+1) for t in range(1,PE+1))
prob += TempWarehouseCost == lpSum(PO * OX[t] for t in range(1,PE+1))
# 供应商权重约束
prob += TotalWeight == lpSum(W[i-1] * X[i,j,t]  for i in range(1,M+1) for j in range(1,PR+1) for t in range(1,PE+1))

# 供应链总信任度约束
prob += TotalReliability == lpSum(Rs[i-1][t-1] * X[i,j,t]  for i in range(1,M+1) for j in range(1,PR+1) for t in range(1,PE+1)) + lpSum(Rw[t-1] * XK[k,j,t] for k in range(1,KR+1) for j in range(1,PR+1) for t in range(1,PE+1) )

# (5) Inventory balance concept. 
# 期初库存I[j,t=0] = 0 
for j in range(1,PR+1):
    prob += I[j,0] == 0
    for t in range(1,PE+1):
        prob += lpSum(X[i,j,t] for i in range(1,M+1)) + I[j,t-1] - lpSum(XK[k,j,t] for k in range(1,KR+1)) == I[j,t]

# (6) The demand for each product at each period
for j in range(1,PR+1):
    for t in range(1,PE+1):
        prob += D[j,t] == lpSum(DK[k-1][j-1][t-1] * B[k,j,t] for k in range(1,KR+1))
        
# （9）中央仓库接收的次品率限制
for j in range(1,PR+1):
    for t in range(1,PE+1):
        prob += lpSum( X[i,j,t] * qa[i-1] for i in range(1,M+1)) <= Qa[j-1] * D[j,t]

# (10)从供货商的订货量不超过供货商的供应能力
for i in range(1,M+1) :
    for j in range(1,PR+1) :
        for t in range(1,PE+1):
            prob += X[i,j,t] <= C[i-1][j-1][t-1]

# (11)从各供货商拿货的总时间不超过最大时间限制
for j in range(1,PR+1) :
    for t in range(1,PE+1):
        prob += lpSum(st[i-1][j-1] * Y[i,j,t] for i in range(1,M+1)) <= LT[j-1][t-1]
# (12) 从供货商进货的总量不超过中央仓库和临时仓库的存储量限制
for j in range(1,PR+1) :
    for t in range(1,PE+1):
        prob += lpSum(X[i,j,t] for i in range(1,M+1))+ I[j,t-1] <= SC[j-1][t-1] + SO[j-1]*OX[t]

# (13)批发商的需求 = 中央仓库的供应量 + 销售损失
for j in range(1,PR+1):
    for k in range(1,KR+1):
        for t in range(1,PE+1):
            prob += XK[k,j,t] + LO1[k,j,t] == DK[k-1][j-1][t-1]
# （15）临时仓库的最大数量
for t in range(1,PE+1):
    prob += OX[t] <= UR
    
# (16) 是否选择从供货商i处进货
for i in range(1,M+1) :
    for j in range(1,PR+1) :
        for t in range(1,PE+1):
            prob += X[i,j,t] <= 1000000*Y[i,j,t]

# 求解问题
prob.solve()

if LpStatus[prob.status] == "Optimal":  
    print("找到最优解")
else:
    print("No solution")
```
